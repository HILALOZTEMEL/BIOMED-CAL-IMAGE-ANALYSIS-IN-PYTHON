**Loading images**

* imageio : Fotoğrafı okur ve kaydeder.
* imread : Tek bir dosya olarak alır ve onu görüntü nesnesi olarak yükler. 
---
imageio ' nun kullanışlı olmasının sebebi ,insan tıbbi görüntülemesi için
standart formatta olan DICOM dosyalarını okuyabilir. Veriler Numpy dizisi türünde görüntü nesnesi olarak okunur.Belirli nesnelere erişebilmek için görüntü index dilimlerine ayrılabilir.


`
im[0:2,0:2]
`




```python
import imageio

im=imageio.imread("body.dcm")
type(im)
```

**Meta data**

Görüntüler her zaman belirli bir bağlamda elde edilir.Bu bilgilere genellik meta veri denir.imageio mevcut mete varilerini meta özelliği aracılığıyla erişilebilen bir sözlüğe yükler. 

`
im.meta
`

Bu özellikler çok sayıda hasta ve alım bilgisi içeren DICOM dosyaları için önemlidir.


> meta bir python sözlüğü olduğundan , belirli bilgilere mevcut anahtar verileri ile indexleyerek erişilebilir. 

`
im.meta["Modality"]
`


> Mevcut bütün verileri listelemek için keys yöntemi ile çağırılabilir.


```python
im.meta
```

```python
im.meta["Modality"]
```

```python
im meta.keys()
```

**Plotting images**
Görselleştirme , görüntü analizi için temeldir.

2D görüntünün basit bir grafiğini çizmak için önce matplotlib 'den pyplot modülünü içe aktarın. 

```python
import matplotlib.pyplot as plt
plt.imshow(im,cmap='gray') # cmap='gray' görüntüyü griye çevirme

#dağınıklığı azaltmak için , pyplot nokta ekseni kapalıyken eksen işaretlerini , etiketleri ve çerçeveyi kapatmak yaygındır. 
plt.axis ('off')
plt.show()
```
